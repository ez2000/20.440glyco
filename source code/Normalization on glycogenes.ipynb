```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as plt
import torch as pt
import csv
import pickle #to save notebook at sessions

#set path for pickles to be saved in
pickle_path = '/Users/erikazhang/Dropbox (MIT)/20.440 Biological Networks/project/python pickles/'
```

```python
# open via: 
#load updated df from pickle
pickle_in = open(pickle_path +"glycosorted_TIL.pkl","rb")
glycosorted_TIL = pickle.load(pickle_in)

pickle_in = open(pickle_path +"glycosorted_LN.pkl","rb")
glycosorted_LN = pickle.load(pickle_in)
```

```python
glycosorted_TILv2 = glycosorted_TIL.transpose()
glycosorted_LNv2 = glycosorted_LN.transpose()
```

## DO Z-transform

```python
from sklearn.preprocessing import StandardScaler
```

```python
### TILs ###
toy_df = glycosorted_TILv2.iloc[:-3]
cols_to_standardize = toy_df.columns
scaler = StandardScaler()
toy_df_TIL = toy_df.copy()
toy_df_TIL[cols_to_standardize] = scaler.fit_transform(toy_df[cols_to_standardize])
```

```python
### LN ###
toy_df = glycosorted_LNv2.iloc[:-3]
cols_to_standardize = toy_df.columns
scaler = StandardScaler()
toy_df_LN = toy_df.copy()
toy_df_LN[cols_to_standardize] = scaler.fit_transform(toy_df[cols_to_standardize])
```

```python
## extract biotin, t-cell type, and L-PHA data from original dataframe and append to transformed data
TILscores_to_append = glycosorted_TILv2.iloc[-3:]
LNscores_to_append = glycosorted_LNv2.iloc[-3:]

Ztrans_TIL = pd.concat([toy_df_TIL, TILscores_to_append])
Ztrans_TIL = Ztrans_TIL.transpose()

Ztrans_LN = pd.concat([toy_df_LN, LNscores_to_append])
Ztrans_LN = Ztrans_LN.transpose()
```

```python
Ztrans_TIL.shape
```

```python
# save pickle 
with open(pickle_path + 'Ztrans_TIL.pkl', 'wb') as f:
    pickle.dump(Ztrans_TIL, f)
    
with open(pickle_path + 'Ztrans_LN.pkl', 'wb') as f:
    pickle.dump(Ztrans_LN, f)

f.close()

# # open via: 
# #load updated df from pickle
# pickle_in = open(pickle_path +"Ztrans_TIL.pkl","rb")
# Ztrans_TIL = pickle.load(pickle_in)

# pickle_in = open(pickle_path +"Ztrans_LN.pkl","rb")
# Ztrans_LN = pickle.load(pickle_in)
```
